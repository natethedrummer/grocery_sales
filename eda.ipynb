```python
import pickle as pkl
import xgboost as xgb
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import confusion_matrix, mean_squared_error, roc_curve, auc, accuracy_score, brier_score_loss
from sklearn.calibration import calibration_curve
from matplotlib import pyplot as plt
%matplotlib inline
import shap
shap.initjs()
```

```python
df = pd.read_csv('/data/nford/hackathon/train.csv')

df = df.set_index('id')

cols = ['store_nbr', 'family']

for i in cols:
    df = pd.concat([df, pd.get_dummies(df[i], prefix=i, dummy_na=True, drop_first=True)], axis=1)
    df = df.drop(columns=[i])

df = df.drop(columns=['date'])
```

```python
# create xgboost data objects
y = df['sales']

X = df.drop(columns=['sales'])
cols = X.columns.tolist()

xgb_full = xgb.DMatrix(X, label=y)

rng = np.random.RandomState(7)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=rng)

xgb_train = xgb.DMatrix(X_train, label=y_train)

xgb_test = xgb.DMatrix(X_test, label=y_test)
```

```python
# define booster parameters
optimal_params = {
    'objective':'reg:squarederror',
    'eval_metric':'rmsle',
    'learning_rate': 0.03,
    'colsample_bytree': 0.6,
    'subsample': 0.8,
    'reg_alpha': 0.3,
    'max_depth': 6,
    'gamma': 4
}

# train booster
xgb_model = xgb.train(
    params=optimal_params, 
    dtrain=xgb_train, 
    num_boost_round=1000, 
    evals=[(xgb_test, "test")], 
    verbose_eval=1000
)
```

```python
explainer = shap.TreeExplainer(xgb_model)
shap_values = explainer.shap_values(X)
shap.summary_plot(shap_values, X)
```
